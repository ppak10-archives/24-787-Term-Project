```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
old_data = pd.read_csv('../uci_supercondutor_data/train.csv')
old_data = old_data.drop(columns=['critical_temp'])
old_data
```

```python
old_element = pd.read_csv('../uci_supercondutor_data/unique_m.csv')
old_element
```

```python
old_data_merge = pd.concat([old_data,old_element],axis = 1,join='outer')
old_data_merge.columns
```

```python
new_data = old_data_merge.groupby('material').mean().reset_index()
new_data.to_csv('new_data_merge.csv',index=None)
```

```python
new_data
```

```python
formula_data = pd.DataFrame(new_data,columns=['material','critical_temp']).reset_index()
formula_data.columns=['material_id','composition','target']
```

```python
formula_data.to_csv('material_data.csv',index=None)
```

```python
train,test = train_test_split(formula_data,train_size=0.6)
```

```python
train.to_csv('roost_train.csv',index=None)
```
